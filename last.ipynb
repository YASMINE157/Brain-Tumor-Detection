```python
import gradio as gr
import cv2
import numpy as np

def predict(image):
    # YOLO prediction
    results = model.predict(source=image, imgsz=640, conf=0.25)

    for result in results:
        class_ids = result.boxes.cls.int().tolist()  # List of class IDs
        if len(class_ids):
            boxes = result.boxes.xyxy  # Bounding boxes from the YOLO model
            # SAM model processing using the bounding boxes
            sam_results = sam_model(result.orig_img, bboxes=boxes, verbose=False, save=True, device=0)

    annotated_img = sam_results[0].plot()
    annotated_img = annotated_img[:, :, ::-1]
    return annotated_img
# Gradio Interface for the app
app = gr.Interface(
    fn=predict,
    inputs=gr.Image(type="numpy", label="Upload an image"),
    outputs=gr.Image(type="numpy", label="Detect Brain Tumor"),
    title="Brain Tumor Detection Using Yolov11",
    description="Upload an image and the yolov11 model will detect and annotate brain tumor"
)

# Launch the app
app.launch()
```
